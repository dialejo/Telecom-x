## **Importando los Datos**

```python
import pandas as pd
#importando la data
url_data = '/content/drive/MyDrive/Challenge telecom x/TelecomX_Data.json'
df = pd.read_json(url_data)
```

##**Normalizando el DataFrame**

```python
#Columnas que tienen diccionarios
columnas_anidadas = ['customer','phone','internet','account']
```

```python
#normalizando las columnas con diccionarios
for columns in columnas_anidadas:
  df_anidado = pd.json_normalize(df[columns])
  df = pd.concat([df, df_anidado],axis = 1)

df.drop(columnas_anidadas,axis = 1,inplace = True)
```

##**Eliminando Datos no validos**

```python
# FunciÃ³n para detectar valores que no son numÃ©ricos vÃ¡lidos
def es_numero_valido(x):
    try:
        float(x)
        return True
    except:
        return False

# Filtrar valores no convertibles
valores_problema = df[~df['Charges.Total'].apply(es_numero_valido)]['Charges.Total'].unique()

print("Valores problemÃ¡ticos:")
print(valores_problema)
```

```python
#eliminando posibles errores de string
df['Charges.Total'] = df['Charges.Total'].replace('', np.nan)
df['Charges.Total']= df['Charges.Total'].apply(lambda x: x.replace(',','').replace('$','').strip() if isinstance(x, str) else x)
```

```python
df['Charges.Total'].isna().sum()
```

```python
df['Charges.Total']= df['Charges.Total'].astype(np.float64)
```

```python
df.info()
```

```python
#revisando las filas que el cargo total es 0
filter_nan = df[df['Charges.Total'].isna()]
filter_nan
```

```python
# despues de analizar las filas podemos concluir que como el tenuere es 0 eso significa que el charge total es nan lo podemos poner como cero
df['Charges.Total'] = df['Charges.Total'].fillna(0)
```

```python
#revisando que la data este limpia
filter_nan = df[df['Charges.Total'].isna()]
filter_nan
```

```python
df.isna().sum()
```

```python
# deberiamos tener ya la misma cantidad de datos en todas las columnas
df.info()
```

```python
# revisando valores duplicados, en este caso si el dustomer ID se repite seria un valor duplicado
df.duplicated('customerID').sum()
```

#ðŸ“Š Carga y anÃ¡lisis

##**Churn by tenure**

```python
#agrupando churn by

df_by_tenure = df.groupby('tenure')['Churn'].value_counts().unstack()
df_by_tenure
```

```python
df_by_tenure['Yes'] = df_by_tenure['Yes'].fillna(0)
df_by_tenure
```

```python
df_by_tenure.info()
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_by_tenure_reset = df_by_tenure.reset_index()
df_long = df_by_tenure_reset.melt(id_vars='tenure',
                                  value_vars=['Yes', 'No'],
                                  var_name='Churn',
                                  value_name='Count')
df_long.head()
```

###**Tabla Dinamica Churn**

```python

fig = px.line(df_long, x='tenure', y='Count',color='Churn',
              title='Churn Trend by Tenure (1 to 72 months)',
              markers=True)
fig.update_layout(width = 800, height= 400, xaxis = {'tickangle': -45},
                  xaxis_title='meses',yaxis_title = 'Number of churn(yes)',
                  font_family ='Arial', font_size = 14, font_color = 'grey')
fig.write_html('/content/drive/MyDrive/Graphs/Challenge one/Churn_Trend_by_tenure.html')
fig.show()
```

###**Tabla Churn reporte**

```python
fig, ax = plt.subplots(1,2,figsize=(12, 4))  # axs es un array de ejes

# Primer grÃ¡fico: Churn = Yes
ax[0].plot(df_by_tenure.index, df_by_tenure['Yes'], color='red', label='Churn: Yes', linewidth=2)
ax[0].set_title('Churn: Yes', loc='left', fontsize=14)
ax[0].legend()

ax[1].plot(df_by_tenure.index, df_by_tenure['No'], color='red', label='Churn: Yes', linewidth=2)
ax[1].set_title('Churn: Yes', loc='left', fontsize=14)
ax[1].legend()

ymin = 0
ymax = 420

for ax in ax.flat:
  ax.set(xlabel='Meses', ylabel='Number of Churn')
  ax.grid(True, color = 'lightgrey')
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_ylim(ymin, ymax)

# TÃ­tulo general
fig.suptitle('Churn Trend by Tenure (1 to 72 months)', fontsize=18, ha = 'right')

plt.savefig('/content/drive/MyDrive/Graphs/Challenge one/Churn_Trend_by_tenure.png',transparent=False,dpi=300,bbox_inches='tight')

plt.tight_layout()
plt.show()
```

##**Analisis 0 a 2 meses Tenuere**

```python
df_analisis_02_tenurer = df[(df['tenure']>=0) & (df['Churn']=='Yes') & (df['tenure']<=3)]
df_analisis_02_tenurer.head()
```

```python
conteo_by_payment = df_analisis_02_tenurer['PaymentMethod'].value_counts().reset_index().sort_values(by='count',ascending=False)
conteo_by_payment.head()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=conteo_by_payment,x='count',y='PaymentMethod',orient='h',palette='mako')
ax.set_title('amount of Chur by payment method for the first 3 months',
       loc ='left', fontsize = 16)
ax.set_xlabel('Number of churn by payment method', fontsize = 14)
ax.set_ylabel('')
ax.set_frame_on(False)
ax.tick_params(axis='y',which = 'both', length = 0)
ax.xaxis.set_visible(False)

for i,j in enumerate(conteo_by_payment['count']):
  ax.text(j+2,i,str(j),color='black',fontsize=12,ha='left',va='center')

fig.savefig('/content/drive/MyDrive/Challenge telecom x/Churn_by_payment_method.png',transparent=False,dpi=300,bbox_inches='tight')
plt.show()
```
